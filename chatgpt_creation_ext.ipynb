```python
import gradio as gr
from transformers import GPT2Tokenizer, GPT2LMHeadModel

model_name = "gpt2-large"
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name)

def chat_with_gpt2(user_input, history=[]):
    if not user_input.strip():
        return history, history

    input_ids = tokenizer.encode(user_input, return_tensors="pt")
    response_ids = model.generate(
        input_ids,
        max_length=120,
        num_return_sequences=1,
        temperature=0.8,
        pad_token_id=tokenizer.eos_token_id
    )
    response = tokenizer.decode(response_ids[0], skip_special_tokens=True)

    history.append(("🧑‍💻 " + user_input, "🤖 " + response))
    return history, history

# ------------------------------
# Custom CSS (Dark + White Input)
# ------------------------------
custom_css = """
body {
    background: radial-gradient(circle at top left, #0a0f1f, #1b2735, #090a0f);
    color: #f0f0f0;
    font-family: 'Poppins', sans-serif;
}

.gradio-container {
    background: transparent !important;
    color: #f0f0f0 !important;
}

/* Chatbox area */
.chatbot {
    background: rgba(255, 255, 255, 0.05);
    border-radius: 15px;
    padding: 10px;
    backdrop-filter: blur(8px);
}

/* Input and textarea — Light background with black text */
textarea, input, .svelte-1ipelgc, .svelte-12x80jh {
    background: #ffffff !important;   /* White input box */
    color: #000000 !important;        /* Black text */
    border: 2px solid #00e0ff !important;
    border-radius: 10px !important;
    font-size: 16px !important;
    padding: 8px !important;
    caret-color: #000000 !important;
}

textarea::placeholder, input::placeholder {
    color: #555555 !important;
}

/* Buttons */
button {
    background: linear-gradient(90deg, #06beb6, #48b1bf) !important;
    color: white !important;
    font-weight: bold !important;
    border-radius: 10px !important;
    border: none !important;
    transition: all 0.3s ease !important;
}

button:hover {
    transform: scale(1.05);
    background: linear-gradient(90deg, #48b1bf, #06beb6) !important;
}

/* Title */
h1 {
    color: #00e0ff !important;
    text-shadow: 0px 0px 15px #00e0ff;
}
"""


with gr.Blocks(css=custom_css, theme="gradio/soft") as demo:
    gr.Markdown(
        """
        <h1 style='text-align:center;'>⚡ GPT-2 Chatbot</h1>
        <p style='text-align:center; color:#b0e0ff;'>Modern dark theme • Light input box • Clean layout</p>
        """
    )

    chatbot = gr.Chatbot(label="", elem_classes="chatbot", height=400)
    user_input = gr.Textbox(
        placeholder="Type your message here...",
        label="💬 Your Message",
        lines=2
    )

    with gr.Row():
        send_btn = gr.Button("🚀 Send", scale=1)
        clear_btn = gr.Button("🧹 Clear Chat", scale=1)

    state = gr.State([])

    send_btn.click(chat_with_gpt2, [user_input, state], [chatbot, state])
    user_input.submit(chat_with_gpt2, [user_input, state], [chatbot, state])
    clear_btn.click(lambda: ([], []), None, [chatbot, state])


demo.launch(share=True)
```
