```python
def summerize_text(text):
  model=genai.GenerativeModel(MODEL_NAME)
  prompt=f"summarize the following text : \n\n{text}"
  response=model.generate_content(prompt)
  return response.text
```

```python
long_text="a distinct section of a piece of writing, usually dealing with a single theme and indicated by a new line, indentation, or numbering:"
summary=summerize_text(long_text)
print("original text:\n",long_text)
print("\n output summerize text\n",summary)
```
